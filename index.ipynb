```python
## Case Study: Web Scraping Jumia

url = 'https://www.jumia.com.ng/catalog/?q=laptops'

r = requests.get(url)

soup = BeautifulSoup(r.text, 'lxml')
```

```python
# Product names
product_names = []
names = soup.find_all('h3', class_='name')
for name in names:
    product_names.append(name.string)
```

```python
product_names
```

```python
## Current Price

current_prices = []
prices = soup.find_all('div', class_='prc')
for price in prices:
    current_prices.append(price.string)
```

```python
current_prices
```

```python
## Former Price

former_prices = []
prices = soup.find_all('div', class_='old')
for price in prices:
    former_prices.append(price.string)
```

```python
former_prices
```

```python
## discount percentage

discount_perc = []
discounts = soup.find_all('div', class_='bdg _dsct _sm')
for discount in discounts:
    discount_perc.append(discount.string)

```

```python
discount_perc
```

```python
data = {
        'title': product_names, 
        'current_price': current_prices, 
        'former_prices': former_prices, 
        'discount': discount_perc
        }

laptops_df = pd.DataFrame.from_dict(data, orient='index').T
```

```python
laptops_df
```

```python
# for all the pages

pages = [x for x in range(1, 51)]
product_names = []
current_prices = []
former_prices = []
discount_perc = []

for page in pages:
    url = f'https://www.jumia.com.ng/catalog/?q=laptops&page={page}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')


    # Product names
    names = soup.find_all('h3', class_='name')
    for name in names:
        product_names.append(name.string)

    
    # Current Price    
    prices = soup.find_all('div', class_='prc')
    for price in prices:
        current_prices.append(price.string)

    # Former Price
    prices = soup.find_all('div', class_='old')
    for price in prices:
        former_prices.append(price.string)

        
    # Discount Percentage
    discounts = soup.find_all('div', class_='bdg _dsct _sm')
    for discount in discounts:
        discount_perc.append(discount.string)
```

```python
data = {
        'title': product_names, 
        'current_price': current_prices, 
        'former_prices': former_prices, 
        'discount': discount_perc
        }

laptops_df = pd.DataFrame.from_dict(data, orient='index').T

laptops_df
```

## Data Transformation and Loading


```python
laptops_df.info()
```

```python
laptops_df.fillna({
    'former_prices': '0',
    'discount': '0%'
}, inplace=True)
```

```python
from sqlalchemy import create_engine, text

engine = create_engine("postgresql://postgres:password@localhost:5432/postgres")

with engine.connect() as conn:
    conn.execution_options(isolation_level="AUTOCOMMIT").execute(
        text("CREATE DATABASE Zee_ecommerce")
    )
```

```python
# define database connection parameters

db_params = {
    'host': 'localhost',
    'port': 5432,
    'database': 'zee_ecommerce',
    'username': 'postgres',
    'password': 'password'
}

# define the database connection url with db parameters
db_url = f"postgresql://{db_params['username']}:{db_params['password']}@{db_params['host']}:{db_params['port']}/{db_params['database']}"

# Create the database engine with the  db url
engine = create_engine(db_url)

#Connect to PostgreSQL server

with engine.connect() as connection:
    #create table and load the data into the table
    laptops_df.to_sql('laptops', connection, if_exists='replace', index=False)

print("Data loaded successfully into the laptops table in the Zee_ecommerce database.")
```
